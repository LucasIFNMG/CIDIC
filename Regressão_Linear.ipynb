##Parâmetros



```
#X = Entrada, Y = Saída
X = np.linspace(-10, 10, 100) 
Y = 1.5*X + 4 + np.random.normal(0, 5, len(X))

#Parâmetros da Regressão
a = np.random.normal(0,1,1) 
b = np.random.normal(0,2,1)


#Hiper-Parâmetros
taxaAprendizado = 0.001
num_epocas = 25
estimativa = []
erro = []
```

```

print('a Inicial = ',a)
print('b Inicial = ',b)
```

##Treinamento

```


for j in range(num_epocas):
  estimativa.clear()
  erro.clear()
  print('Época ',j)
  for i in range(len(X)):
    estimativa.append(a*X[i] + b)
    erro.append(Y[i] - estimativa[i])
    a = a - taxaAprendizado * 2 * erro[i] * -X[i]
    b = b - taxaAprendizado * -2 * erro[i]
    
    #Acompanhando os Parâmetros
    if i%50 == 0:
        
        print("i = ",i)
        print('X[i] = %.5f'%(X[i]) ) 
        print('Y[i] = %.5f'%(Y[i]) )
        print('Estimativa = %.5f'%(estimativa[i]))
        print('Erro[i] = %.5f'%(erro[i]))
        print('a = ',a)
        print('b = ',b)
        print('********************')
```

##Gráficos

```
estimativa.clear()
erro.clear()

#PLOTS
#Gráfico 1: X vs Y
plt.plot(X,Y)
plt.title('X vs Y') 
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

for i in range(len(X)):
  estimativa.append(a*X[i] + b) 
  erro.append(Y[i] - estimativa[i])

#Gráfico 2: X vs Estimativa
plt.plot(X, estimativa, color='red')
plt.plot(X,Y)
plt.title('Vermelho: X vs Estimativa | Azul: X vs Y')
plt.xlabel('X')
plt.ylabel('Estimativa')
plt.show()  
```
