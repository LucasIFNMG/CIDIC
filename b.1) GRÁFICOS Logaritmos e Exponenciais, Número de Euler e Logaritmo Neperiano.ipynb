```python
#Logaritmos
#np.log2
#np.log10
import matplotlib.pyplot as plt
import numpy as np
x = arange(1.0,100.0,1.0)
y = np.log2(x)

plot(x,y)
xlabel('x')
ylabel('y')
title('Logaritmo')
grid(True)
 
show()
```

```python
#Logaritmo Neperiano
#np.log2
#np.log10
import matplotlib.pyplot as plt
import numpy as np
x = arange(1.0,100.0,1.0)
y = np.log10(x)

plot(x,y)
xlabel('x')
ylabel('y')
title('Logaritmo Neperiano')
grid(True)
 
show()
```

```python
#Exponencial (Nº de Euler)
from pylab import *
import matplotlib.pyplot as plt
import math
for x in range(0, 11):
        print ('exp %d = %f' % (x,  math.exp(x)))
x = arange(0.0,10.0,0.2)
y = exp(x)

plot(x,y)
xlabel('x')
ylabel('y')
title('Exponencial')
grid(True)
 
show()
```
