```python
#Seno
from pylab import *
 
t = arange(0.0, 5.0,0.01)
y = sin(2*pi*t)
plot(t, y)
 
xlabel('x')
ylabel('sin(x)')
title('Função Seno')
grid(True)
 
show()
```

```python
#Cosseno
from pylab import *
 
t = arange(0.0, 5.0,0.01)
y = cos(2*pi*t)
plot(t, y)
 
xlabel('x')
ylabel('cos(x)')
title('Função Cosseno')
grid(True)
 
show()
```
